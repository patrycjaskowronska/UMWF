```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("creditcard.csv")
df
```

```python
df['Class'].value_counts()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
```

```python
knn_acc = accuracy_score(y_test, y_pred)
print("Accuracy: {}".format(knn_acc))

knn_pr = precision_score(y_test, y_pred)
print("Precision: {}".format(knn_pr))

knn_rec = recall_score(y_test, y_pred)
print("Recall: {}".format(knn_rec))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Undersampling
rus=RandomUnderSampler()
X_train_rus, y_train_rus=rus.fit_resample(X_train, y_train)
y_train_rus.value_counts()
```

```python
knn_rus = KNeighborsClassifier(n_neighbors=3)
knn_rus.fit(X_train_rus, y_train_rus)
y_pred = knn_rus.predict(X_test)
```

```python
knn_rus_acc = accuracy_score(y_test, y_pred)
print("Accuracy: {}".format(knn_rus_acc))

knn_rus_pr = precision_score(y_test, y_pred)
print("Precision: {}".format(knn_rus_pr))

knn_rus_rec = recall_score(y_test, y_pred)
print("Recall: {}".format(knn_rus_rec))
```

```python
# Oversampling
oversample = SMOTE()
X_train_os, y_train_os = oversample.fit_resample(X_train, y_train)
y_train_os.value_counts()
```

```python
knn_os = KNeighborsClassifier(n_neighbors=3)
knn_os.fit(X_train_os, y_train_os)
y_pred = knn_os.predict(X_test)
```

```python
knn_os_acc = accuracy_score(y_test, y_pred)
print("Accuracy: {}".format(knn_os_acc))

knn_os_pr = precision_score(y_test, y_pred)
print("Precision: {}".format(knn_os_pr))

knn_os_rec = recall_score(y_test, y_pred)
print("Recall: {}".format(knn_os_rec))
```
